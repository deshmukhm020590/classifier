```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pathlib import Path

# Project root is two levels up from this notebook (.scripts -> classifier -> project)
PROJECT_ROOT = Path().resolve().parents[1]
DATA_DIR = PROJECT_ROOT / "data"

def load_data_files(data_dir=DATA_DIR):
    """
    Reads all CSV and Excel files from data_dir into a single pandas DataFrame.
    Files are concatenated with a 'source_file' column indicating the origin.
    Returns an empty DataFrame if no files are found.
    """
    readers = {
        ".csv":  lambda p: pd.read_csv(p),
        ".xlsx": lambda p: pd.read_excel(p),
        ".xls":  lambda p: pd.read_excel(p),
    }

    frames = []
    for path in sorted(Path(data_dir).iterdir()):
        reader = readers.get(path.suffix.lower())
        if reader:
            df = reader(path)
            df["source_file"] = path.name
            frames.append(df)
            print(f"Loaded: {path.name}  ({len(df)} rows)")

    if not frames:
        print(f"No CSV/Excel files found in: {data_dir}")
        return pd.DataFrame()

    return pd.concat(frames, ignore_index=True)

df = load_data_files()
df.head()
```
